```python
# from bs4 module using BeautifulSoup and requests library

from bs4 import BeautifulSoup
import requests
```

```python
# the url to the wesbsite from where we wanted to extract the data
url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'

page = requests.get(url) # link the url to the page.
```

```python
soup = BeautifulSoup(page.text, 'html') # using beautifulSoup to parse the text which is in HTML lang.

#print(soup)  #Entire HTML script. you can also use prettify to format it a little bit.
#print(soup.prettify)
```

```python
tableData = soup.find_all('table', class_='table')
#print(tableData)
```

```python
# Extraxting the table headers

title = soup.find_all('th') # we got every headers for this table.


tableHeader = [title.text.strip() for title in title] # creating a list of headers

# there are some extra headers which i do not need from index 4.
n=0
column_headers=[] # empty list

for list in tableHeader:
    column_headers.append(list) # appending the empty list in the headers.
    n=n+1
    if n==4:
        break
```

```python
# import pandas library to manipulate and create the file
import pandas as pd
```

```python
df = pd.DataFrame(columns = column_headers) # store the Headers in the data frame
```

```python
# extracting the table Data
columnData=soup.find_all('tr', class_='tr')[0:] # found every table row

for row in columnData: # running loop through each table row 
    row_data = row.find_all('td') #found every row data
    individual_row = [row_data.text.strip() for row_data in row_data] # after clearing the format of each row, we are assigning it to the individual row
    #print(individual_row)
    length = len(df) #check the length of the dataframe
    df.loc[length] = individual_row; # store the data in each row in the data frame
```
