```python
import pandas as pd
import numpy as np # used for many mathmetical operations such as trignometrics, statistics and linear algebra.
import matplotlib.pyplot as plt # used for plotting/visualization creating varity of static, animated and interactive plots to visualize data.
```

```python
df = pd.read_csv(r"C:\Users\GAURAV\Desktop\Data Analytics\Python Workshop\Pandas library\Ice Cream Ratings.csv",index_col = 'Date')
df
```

```python
# inbuilt visualization using matplotlib library

print(plt.style.available) # you can print all the style that are avaialble
plt.style.use('Solarize_Light2') # it will use same style throught the chart that you will create.
```

```python
# using plot we can create graph, 
# kind = means which graph you want be default it is line you can specify bar, hist, scatter, etc
# xlabel = labelling for x-axis
# ylabel  = label for y-axis
#subplots = If true, create separate cha"rts.

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores', subplots = False)
```

```python
# stacked = True created stacked bar chart.
# if kind  = barh, create horizantal bar chart by default it is vertical bar chart.
# only take flavor rating column
df['Flavor Rating'].plot(kind = 'bar', stacked = False, xlabel = 'Ratings', ylabel = 'Scores')
```

```python
#Horizantal bar chart
df.plot.barh(stacked = 'True')
```

```python
# scatter plot
# s means size
# c means color
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 450, c = 'Purple')
```

```python
# histogram 
# bins refer to the width of column higher number means less width
df.plot.hist(bins = 80)
```

```python
df.boxplot()
```

```python
# area chart
# fig size = (width, height)
df.plot.area(figsize = (6,3))
```

```python
# pie chart
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
